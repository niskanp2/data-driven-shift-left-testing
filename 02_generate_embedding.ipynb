```python
import pickle
import networkx as nx
from karateclub import Role2Vec
import json

from src.ml_models import DIMENSIONS
```

```python
with open("gen/network.pickle", "rb") as f:
    G = pickle.load(f)

G_emb = nx.convert_node_labels_to_integers(G, label_attribute="name")
node_label_to_idx = {data["name"]: node for node, data in G_emb.nodes(data=True)}
node_idx_to_label = {node: data["name"] for node, data in G_emb.nodes(data=True)}
```

```python
r2v = Role2Vec(
    walk_number=200,
    walk_length=4,
    window_size=2,
    dimensions=DIMENSIONS,
    workers=4,
    min_count=1,
)

r2v.fit(G_emb)
embeddings = r2v.get_embedding()
```

```python
embedding_dict = {}
for i in range(len(embeddings)):
    embedding_dict[node_idx_to_label[i]] = [
        float(embedding) for embedding in embeddings[i]
    ]

with open("gen/embeddings.json", "w") as f:
    json.dump(embedding_dict, f)
```
